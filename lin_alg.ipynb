```python
import numpy as np 
# Checking linear independence of vectors using NumPy
X = np.array([
    [1, 1, -1],  # x1
    [2, 1, -2],  # x2
    [-3, 0, 1],  # x3
    [4, 2, 1]    # x4
]).T  # Each column = one vector  
rank = np.linalg.matrix_rank(X)      
print("Rank of the matrix:", rank)   
print("Linear Independent:" if rank == X.shape[1] else "Linear Dependent") 
```

```python
import numpy as np
from scipy.linalg import qr

# Compute QR with column pivoting
Q, R, pivot = qr(X, mode="economic", pivoting=True)

# Compute rank from R
tol = 1e-12
rank = np.sum(np.abs(np.diag(R)) > tol)

# Get basis columns
basis_cols = pivot[:rank]

# Extract the basis vectors (columns)
basis_vectors = X[:, basis_cols]

print(basis_vectors)
```
